```python
import os

benign_dir = 'C:\Users\pragy\Desktop\ibm\inbreast\benign'
malignant_dir = 'C:\Users\pragy\Desktop\ibm\inbreast\malignant'

benign_count = len(os.listdir(benign_dir))
malignant_count = len(os.listdir(malignant_dir))

print(f"Benign: {benign_count}, Malignant: {malignant_count}")
```

```python
%pip install torchvision

from torchvision import transforms
from PIL import Image
import random
import shutil
```

```python
augment = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.1, contrast=0.1),
    transforms.RandomResizedCrop(size=224, scale=(0.9, 1.0)),
    transforms.ToTensor()
])

# Directory for augmented images
augmented_benign_dir = 'C:\Users\pragy\Desktop\ibm\inbreast\benign_augmented'
os.makedirs(augmented_benign_dir, exist_ok=True)

benign_images = os.listdir(benign_dir)
augment_needed = malignant_count - benign_count

for i in range(augment_needed):
    img_name = random.choice(benign_images)
    img_path = os.path.join(benign_dir, img_name)
    image = Image.open(img_path).convert('RGB')
    aug_image = augment(image)
    # Save augmented image
    save_path = os.path.join(augmented_benign_dir, f"aug_{i}_{img_name}")
    transforms.ToPILImage()(aug_image).save(save_path)
```
