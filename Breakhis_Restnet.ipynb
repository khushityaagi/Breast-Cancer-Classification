```python
import os
import shutil

input_root = "/content/breakhis/BreaKHis_v1/BreaKHis_v1/histology_slides/breast"
output_root = "organized_dataset/train"
magnification = "100X"

for label in ["benign", "malignant"]:
    sob_root = os.path.join(input_root, label, "SOB")
    if not os.path.exists(sob_root):
        continue

    for subtype in os.listdir(sob_root):
        subtype_path = os.path.join(sob_root, subtype)
        if not os.path.isdir(subtype_path):
            continue

        for case_id in os.listdir(subtype_path):
            case_path = os.path.join(subtype_path, case_id, magnification)
            if not os.path.exists(case_path):
                continue

            for img in os.listdir(case_path):
                src = os.path.join(case_path, img)
                dst_dir = os.path.join(output_root, label)
                os.makedirs(dst_dir, exist_ok=True)
                dst = os.path.join(dst_dir, f"{subtype}_{case_id}_{img}")
                shutil.copyfile(src, dst)

print(" All 100X images copied successfully.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_size = (224, 224)
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    "data_split/train",
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_gen = val_datagen.flow_from_directory(
    "data_split/val",
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # freeze pretrained layers

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # binary classification
])

model.compile(optimizer=Adam(1e-4), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(patience=3, restore_best_weights=True),
    ModelCheckpoint("best_breakhis_resnet100x.h5", save_best_only=True)
]

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,
    callbacks=callbacks
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.legend()
plt.title("ResNet50 Accuracy (100X)")
plt.show()
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Step 1: Enhanced Data Augmentation
img_size = (224, 224)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    "data_split/train",
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_gen = val_datagen.flow_from_directory(
    "data_split/val",
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False  # for consistent evaluation
)
```

```python
# Step 2: Build & Fine-tune ResNet
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = True  # fine-tuning enabled

# Freeze first few layers to prevent overfitting
for layer in base_model.layers[:-50]:
    layer.trainable = False

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Step 3: Add class weights to balance bias
from sklearn.utils import class_weight
import numpy as np

val_gen.reset()
y_true = val_gen.classes

class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_true),
    y=y_true
)

class_weights_dict = dict(enumerate(class_weights))
print("Class Weights:", class_weights_dict)
```

```python
# Step 4: Train with class weights
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(patience=3, restore_best_weights=True),
    ModelCheckpoint("resnet_trial2_best.h5", save_best_only=True)
]

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,
    class_weight=class_weights_dict,
    callbacks=callbacks
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict on validation set
val_gen.reset()
y_pred_probs = model.predict(val_gen)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()
y_true = val_gen.classes

# Classification Report
print(" Classification Report (Trial 2):\n")
print(classification_report(y_true, y_pred, target_names=val_gen.class_indices.keys()))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
labels = list(val_gen.class_indices.keys())

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Trial 2")
plt.show()
```
